```python
import pandas as pd
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
a1=sns.histplot(tips['total_bill']) # histplot 히스토그램
```

```python
a1=sns.kdeplot(tips['total_bill']) # 밀도함수그래프
```

```python
a1=sns.kdeplot(tips['total_bill'],tips['tip']) # 밀도함수그래프
```

```python
a1=sns.kdeplot(tips['total_bill'],tips['tip'],shade=True)
# shade=True 로 밀도 높은 곳은 색깔이 진하게
```

```python
a1=sns.countplot(tips['day']) # 특정 값 계수
```

```python
a1=sns.lmplot(x='total_bill',y='tip',data=tips) # 산점도
```

```python
a1=sns.jointplot(x='total_bill',y='tip',data=tips) # 산점도 + 히스토그램
```

```python
a1=sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
# kind='hex'를 통해 산점도의 밀도가 높을수록 색깔을 진하게 나타냄
```

```python
a1=sns.barplot(x='time',y='total_bill',data=tips) # 평균 막대그래프
# 검은 선은 신뢰구간을 나타냄
```

```python
a1=sns.boxplot(x='time',y='total_bill',data=tips) # 상자그래프
# 색깔박스 안에 있는 선은 중앙값 median임(50%라는 뜻인듯?)
# 색깔 박스의 윗면은 상위 25퍼, 아랫면은 상위 75퍼
# 색깔 박스 위의 울타리는 upper fence로 색깔박스 안에 있는 값들의 1.5배에 해당하는 범위
# upper fence까지가 통계적으로 봤을때 신뢰할 수 있다는 뜻.
# 대부분의 값은 상자안에 있다는 의미
```

```python
a1=sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
a1=sns.violinplot(x='time',y='tip',data=tips) # 바이올린 그래프
# boxplot이 밀집도를 확인하기 어려운 걸 보완하기 위해 만듬
```

```python
a1=sns.pairplot(tips) # 관계그래프
# 비교할 수 있는 모든 열을 비교해서 나타내줌
```

```python
a1=sns.pairplot(tips, hue='sex')
```

```python
sns.violinplot(x='time',y='total_bill',data=tips,hue='sex')
```

```python
sns.violinplot(x='time',y='total_bill',data=tips,hue='sex',split=True)
```

```python
anscombe=sns.load_dataset('anscombe')
sns.lmplot(x='x',y='y',data=anscombe, col='dataset')
# 'dataset' 별로 나눠서 그림
```

```python
anscombe=sns.load_dataset('anscombe')
sns.lmplot(x='x',y='y',data=anscombe,col='dataset',col_wrap=2)
# 최대 2열까지 표시
```

```python
import matplotlib.pyplot as plt
A=sns.FacetGrid(tips, col="time")
A.map(plt.scatter,"total_bill",'tip')
```

```python
A=sns.FacetGrid(tips, col="time",hue='sex')
A.map(plt.scatter,"total_bill",'tip')
```

```python
sns.set_style('whitegrid') # 배경
A=sns.FacetGrid(tips, col="time",hue='sex')
A.map(plt.scatter,"total_bill",'tip')
```

```python
sns.set_style('whitegrid')
A=sns.FacetGrid(tips, col="time",hue='sex',row='smoker') # col: 열, row: 행
A.map(plt.scatter,"total_bill",'tip')
```
