```python
import pandas as pd

ebola=pd.read_csv('/content/country_timeseries.csv')
ebola
```

```python
ebola.info()
# 데이터의 총개수는 122개라는 뜻
# date 열은 122개여서 누락값 없음
# cases_Guinea의 경우 데이터가 93개 나머지는 누락값
```

```python
122-ebola.count()
# 누락값 갯수
```

```python
ebola.fillna(5)
```

```python
ebola.mean()
# 각 열들의 평균 값
```

```python
ebola.fillna(ebola.mean())
# 각 열의 누락값들에 대해 각 열의 평균값을 넣음
```

```python
ebola.fillna(method='ffill')
# 결측치를 앞에 있는 값으로 채움
# 앞에 있는 데이터가 없어서 NaN이 나오기도 함
```

```python
ebola.fillna(method='bfill')
# 결측치를 뒤에 있는 값으로 채움
# 뒤에 있는 데이터가 없어서 NaN이 나오기도 함
```

```python
ebola.interpolate()
# 선형 흐름에 따라 결측치 채움
# 컴퓨터가 흐름을 알 수 없는 것들에 대해서는 NaN 처리
```

```python
ebola2=ebola.interpolate()
ebola2=ebola2.fillna(method='ffill')
ebola2=ebola2.fillna(method='bfill')
ebola2
```

```python
dates=['12/1/2016','12/3/2016','12/4/2016','12/10/2016']
dates=pd.to_datetime(dates)
# pandas에서 제공하는 시간데이터 변환 함수
dates
```

```python
from numpy import nan
```

```python
s=pd.Series([1,nan,nan,10],index=dates)
s
```

```python
s.interpolate()
# 1일에 1, 2일에 2, 3일에 3 이어야하는데 컴퓨터는 선형흐름으로만 채움
```

```python
s.interpolate(method='time')
# 선형흐름이 아닌 시간에 따라 데이터를 채우라고 지시
```
