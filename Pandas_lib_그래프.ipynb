```python
import pandas as pd
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
tips['total_bill'].plot()
```

```python
tips['total_bill'].plot.hist() # 히스토그램
```

```python
tips[['total_bill','tip','size']].plot.hist(alpha=0.5,bins=30)
# alpha: 투명도, bins: 막대개수
```

```python
tips[['tip']].plot.kde() # 아래면적의 넓이는 1(밀도이므로)
```

```python
tips[['total_bill','tip']].plot.kde()
```

```python
tips.plot.scatter(x='total_bill',y='tip')
```

```python
tips.plot.scatter(x='total_bill',y='tip',c='sex') # color로 구분
```

```python
tips.plot.scatter(x='total_bill',y='tip',c='size')
```

```python
tips.plot.scatter(x='total_bill',y='tip',c='total_bill',s=tips['size']*30) # 크기로 구분
```

```python
tips.plot.hexbin(x='total_bill',y='tip')
```

```python
tips.plot.hexbin(x='total_bill',y='tip',gridsize=20) # gridsize 육각형 크기(작을수록 크기가 커짐)
```

```python
tips.plot.box() # 숫자로된 데이터만 나옴
```

```python
tips.plot.pie(x='sex',y='size')
```
