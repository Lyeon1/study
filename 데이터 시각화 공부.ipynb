```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
#앤스컴 : 평균, 표준편차, 상관계수, 회귀계수가 같은 연습용 데이터
A=sns.load_dataset('anscombe')
A
```

```python
a1=A[A['dataset']=='I']
a2=A[A['dataset']=='II']
a3=A[A['dataset']=='III']
a4=A[A['dataset']=='IV']
```

```python
#평균
print(a1.mean())
print(a2.mean())
print(a3.mean())
print(a4.mean())
```

```python
#표준편차
print(a1.std())
print(a2.std())
print(a3.std())
print(a4.std())
```

```python
B=plt.figure()
b1=B.add_subplot(2,2,1)
b2=B.add_subplot(2,2,2)
b3=B.add_subplot(2,2,3)
b4=B.add_subplot(2,2,4)
```

```python
# scatter, 값을 점찍어줌
b1.scatter(a1['x'],a1['y']) # x축은 a1의 x열, y축은 a1의 y열
b2.scatter(a2['x'],a2['y'])
b3.scatter(a3['x'],a3['y'])
b4.scatter(a4['x'],a4['y'])
B
```

```python
# plot, 선그래프
b1.plot(a1['x'],a1['y'], 'red')
b2.plot(a2['x'],a2['y'], 'red')
b3.plot(a3['x'],a3['y'], 'red')
b4.plot(a4['x'],a4['y'], 'red')
B
```

```python
b1.set_title("Data 1")
b2.set_title("Data 2")
b3.set_title("Data 3")
b4.set_title("Data 4")
B
```
